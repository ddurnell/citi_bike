### Imports

```python
# Imports
import warnings; warnings.simplefilter('ignore')
import pandas as pd
import os
import glob
```

### Loop through the files and make them one

```python
# Go to the .csv directory
os.chdir("../data")
```

```python
# Do some regex matching
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
# Combine all files in the list with pandas
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames])
#export to csv
combined_csv.to_csv("combined_data.csv", index=False, encoding='utf-8-sig')
```
